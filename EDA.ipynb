<h1>Finding Outliers</h1>

```python
sns.boxplot(df['sale_price'], orient = 'h')
# Here in sale_price we had a lot of outliers
```

```python
# Finding sale_price outliers

iqr = np.quantile(df['sale_price'], 0.75) - np.quantile(df['sale_price'], 0.25)

lower_th = np.quantile(df['sale_price'], 0.25) - (1.5 * iqr)
upper_th = np.quantile(df['sale_price'], 0.75) + (1.5 * iqr)

outliers = df[(df['sale_price'] < lower_th) | (df['sale_price'] > upper_th)]
outliers
# So, we have 1447 outliers in sale_price column
```

```python
# What is the mean and median sale_price in outliers?

outliers['sale_price'].mean() #1208.5134277816173
outliers['sale_price'].median() #940.0
```

```python
# Delete the outliers

df = df.drop(outliers.index)
df
```

```python
sns.boxplot(df['sale_price'])
# Now, our boxplot looks much more better
# We can still see some outliers, but it's because column sale_price isn't normally distributed
```
